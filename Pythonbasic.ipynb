Calculater with function and if, else , elif


```python
def calculator( a, b , operation):
  if operation == "add":
    return a + b
  elif operation == "sub":
    return a - b
  elif operation == "mul":
    return a * b
  elif operation == "div":
    return a / b
  else:
    return ("Invalid operation")

print(calculator(10,20,"add"))
print(calculator(10,20,"sub"))
print(calculator(10,20,"mul"))
print(calculator(10,20,"div"))
print(calculator(10,20,"mod"))
```

Error Handling in API Request

```python
import requests

# API Details
url = "https://chatify.najam.pk/api/v1/sendmessage"
params = {
    "number": "03125250255",
    "message": "Hello, this is a test message",
    "api key": "your_api_key"
}

try:
  response = requests.get(url, params=params)

  if response.status_code == 400:
    print("Message sent successfully")
  else:
    print(f"Failed to send message. Status code: {response.status_code}")
except requests.exceptions.RequestException as e:
  print(f"An error occurred: {e}")
  print(response.status_code)
```

```python
# This is a class
class House:
  address: str
  num_of_doors: int = 4
  num_of_rooms: int = 5

  def __init__ (self,address:str,colour:str):
    self.address = address
    self.colour = colour
  def ring_bell(self):
    print("Ding Dong!")

# This is an object

Sunny_house = House("house no.123","white")
print(f"Sunny House address is: {Sunny_house.address} and its colour is: {Sunny_house.colour}")
Sunny_house.ring_bell()


Qureshi_house = House("house no.471", "half white")
print(f"Qureshi_house address is: {Qureshi_house.address} and its colour is: {Qureshi_house.colour}")
```
